### Let's use the decision tree regression algorithm to train a car price prediction model. So let’s split the data into training and test sets and use the decision tree regression algorithm to train the model:

```python
# obtaining features and labels
# considering only numerical columns
df1 = df[["symboling", "wheelbase", "carlength", 
             "carwidth", "carheight", "curbweight", 
             "enginesize", "boreratio", "stroke", 
             "compressionratio", "horsepower", "peakrpm", 
             "citympg", "highwaympg", "price"]]

X = df1.iloc[:, :-1]
y=df1.iloc[:, -1]
```

```python
# splitting dataset into train and test data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor()
model.fit(X_train, y_train)
```

```python
# checking model's accuracy and mean_squared_error value on test data
from sklearn.metrics import mean_absolute_error
y_pred = model.predict(X_test)
mse = mean_absolute_error(y_test, y_pred)
print(mse)
score = model.score(X_test, y_pred)
print(score)
```

### The model gives 100% accuracy on the test set, which is excellent.
